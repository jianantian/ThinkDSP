```python
import copy
import random
from abc import ABCMeta, abstractmethod
from typing import Callable

import math
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.io import wavfile


COLOR_PALETTE = sns.color_palette('hls', 10)


class Wave:
    """
    Represents a discrete-time waveform.
    """

    def __init__(self, ys, ts=None, framerate: int = 11025):
        """
        Initializes the wave.

        :param ys: wave array
        :param ts: array of times
        :param framerate: samples per second
        """
        self.ys = np.asanyarray(ys)
        self.framerate = framerate
        self.step = 1 / self.framerate

        if ts is None:
            self.ts = np.arange(0, len(ys)) * self.step
        else:
            self.ts = np.asanyarray(ts)

    @classmethod
    def read(cls, filename):
        """

        :param filename:
        :return:
        """
        framerate, ys = wavfile.read(filename)
        return cls(ys, framerate=framerate)

    def write(self, filename):
        """

        :param filename:
        :return:
        """
        max_value = np.max(np.abs(self.ys))
        data = self.ys.astype(np.float32)
        if max_value > 1:
            data = (data / max_value).astype(np.float32)
        wavfile.write(filename, self.framerate, data)

    def __len__(self):
        return len(self.ys)

    def copy(self):
        """
        Makes a copy.

        Returns: new Wave
        """
        return copy.deepcopy(self)

    @property
    def start(self) -> float:
        return self.ts[0]

    @property
    def end(self) -> float:
        return self.ts[-1]

    @property
    def duration(self) -> float:
        """
        Duration (property).

        :returns: float duration in seconds
        """
        return len(self.ys) / self.framerate

    def show(self, color=None):
        """

        :return:
        """
        # fig = plt.figure()
        if color is None:
            num = random.randint(0, 9)
            color = COLOR_PALETTE[num]
        elif isinstance(color, int):
            color = COLOR_PALETTE[color % 10]
        else:
            raise ValueError
        plt.plot(self.ts, self.ys, color=color)
        plt.show()
        # fig.close()


class Signal(metaclass=ABCMeta):
    """
    Represents a time-varying signal.
    """

    @property
    def frequency(self) -> float:
        """

        :return:
        """
        return 440

    @property
    def amplitude(self) -> float:
        """

        :return:
        """
        return 1

    @property
    def framerate(self) -> int:
        """

        :return:
        """
        return 11025

    @framerate.setter
    def framerate(self, value: int) -> int:
        """

        :return:
        """
        pass

    @property
    def period(self) -> float:
        """
        Period of the signal in seconds (property).

        Since this is used primarily for purposes of plotting,
        the default behavior is to return a value, 0.1 seconds,
        that is reasonable for many signals.

        :returns: float seconds
        """
        return 0.1

    @abstractmethod
    def evaluate(self, ts):
        """

        :param ts:
        :return:
        """
        pass

    def make_wave(self, duration: float = 1, start: float = 0, framerate: int = None) -> Wave:
        """
        Makes a Wave object.

        :param duration: float seconds
        :param start: float seconds
        :param framerate: framerate
        returns: Wave
        """
        if framerate is None:
            framerate = self.framerate
        ts = start + np.arange(0, duration, 1/self.framerate)
        ys = self.evaluate(ts)
        return Wave(ys, ts, framerate=framerate)

    def show(self, num_period=3, start=0):
        """
        Plots the signal
        :param num_period:
        :param start:
        :return:
        """
        duration = self.period * num_period
        wave = self.make_wave(duration, start=start)
        wave.show()

    def __add__(self, other: 'Signal' = None) -> 'Signal':
        """
        Adds two signals.

        :param other
        :returns:
        """
        if other is None:
            return self
        return SumSignal(self, other)

    __radd__ = __add__


class SumSignal(Signal):
    """
    Represents the sum of signals.
    """

    def __init__(self, *args: Signal):
        """
        Initializes the sum.

        :param args: tuple of signals
        """
        self.signals = args

    @property
    def period(self) -> float:
        """
        Period of the signal in seconds.

        # todo: this is not correct; it's mostly a placekeeper.
                But it is correct for a harmonic sequence where all component
                frequencies are multiples of the fundamental.

        :returns: float seconds
        """
        return max(sig.period for sig in self.signals)

    def evaluate(self, ts):
        """
        Evaluates the signal at the given times.

        :param ts: float array of times

        :returns: float wave array
        """
        ts = np.asarray(ts)
        return sum(sig.evaluate(ts) for sig in self.signals)


class Sinusoid(Signal):
    """
    """

    def __init__(self,
                 freq: int,
                 amp: float,
                 offset: float = 0,
                 framerate: int = 11025,
                 func: Callable[[float], float] = np.sin):
        """

        :param freq: float frequency in Hz
        :param amp: float amplitude, 1.0 is nominal max
        :param offset: float phase offset in radians
        :param framerate: framerate
        :param func:
        """
        self.__frequency = freq
        self.__amplitude = amp
        self.__offset = offset
        self.__framerate = framerate
        self.__func = func

    @property
    def frequency(self) -> float:
        """

        :return:
        """
        return self.__frequency

    @property
    def amplitude(self) -> float:
        """

        :return:
        """
        return self.__amplitude

    @property
    def framerate(self) -> int:
        """

        :return:
        """
        return self.__framerate

    @framerate.setter
    def framerate(self, value: int):
        """

        :param value:
        :return:
        """
        self.__framerate = value

    @property
    def period(self) -> float:
        """
        Period of the signal in seconds.
        :return:
        """
        return 1.0 / self.__frequency

    def evaluate(self, ts):
        """
        Evaluates the signal at the given times.

        ts: float array of times
        :param ts: float array of times
        :return:
        """
        ts = np.asarray(ts)
        phases = 2 * math.pi * self.__frequency * ts + self.__offset
        ys = self.__amplitude * self.__func(phases)
        return ys


class SinSignal(Sinusoid):
    """

    """

    def __init__(self, freq=440, amp=1.0, offset=0, framerate=11025, func=np.sin):
        """

        :param freq:
        :param amp:
        :param offset:
        :param framerate:
        :param func:
        """
        super().__init__(freq=freq, amp=amp, offset=offset, framerate=framerate, func=func)


class CosSignal(Sinusoid):
    """

    """

    def __init__(self, freq=440, amp=1.0, offset=0, framerate=11025, func=np.cos):
        """

        :param freq:
        :param amp:
        :param offset:
        :param framerate:
        :param func:
        """
        super().__init__(freq=freq, amp=amp, offset=offset, framerate=framerate, func=func)
```
